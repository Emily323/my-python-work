```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### contourf

```python
x =np.sort(np.random.randint(low=1,high=20,size=10))
y = np.sort(np.random.randint(low=1,high=20,size=10))
X,Y = np.meshgrid(x,y)
Z = np.random.randint(1,5,(10,10))
plt.contourf(X,Y,Z)
plt.scatter(X.ravel() , Y.ravel() , c = Z.ravel())
```

```python
x =np.array([2,4,6,8,10])
y = np.array([2,4,6,8,10])
X,Y = np.meshgrid(x,y)
Z = np.random.randint(1,30,(5,5))
plt.contour(X,Y,Z)
plt.scatter(X.ravel() , Y.ravel() , c = Z.ravel())
```
