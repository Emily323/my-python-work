```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from wordcloud import WordCloud
import jieba
from scipy.misc import imread   
```

```python
text = open('C:/Python Work/text.txt','r',encoding="utf-8").read()  
cut_text = ' '.join(jieba.cut(text))
color_mask = imread('C:\Python Work\交作業囉\c.jpg')
cloud = WordCloud(font_path='c:\WINDOWS\FONTS\MSJHL.TTC',
mask=color_mask,
background_color="white",
max_words=50,
max_font_size=32)
word_cloud = cloud.generate(cut_text)
plt.axis('off')
plt.imshow(word_cloud)
plt.show()
```
