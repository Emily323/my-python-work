```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### 台灣外籍學位生的來源

```python
df=pd.read_csv('107.csv')
```

```python
df.drop(['非學位生_外國交換生','非學位生_外國短期研習及個人選讀','非學位生_大專附設華語文中心學生','非學位生_大陸研修生','非學位生_海青班','境外專班'],axis=1,inplace=True)
```

```python
df
```

```python
df.rename(columns={'學位生_僑生(含港澳)':'學位生_僑生'},inplace=True)
```

```python
df['學位生總人數']=df.學位生_正式修讀學位外國生+df.學位生_僑生 +df.學位生_正式修讀學位陸生
```

### 各國家來台學位生總人數排比

```python
df.sort_values(by='學位生總人數',ascending=False) 
```

```python
df_new=df[df.學位生總人數>5000] #畫學位生人數超過五千 國家與人數的圓餅圖
```

```python
df_new
```

```python
Y=df_new.loc[:,'國別']
X=df_new.loc[:,'學位生總人數']


plt.pie(X,labels=Y)
plt.show()
```
