```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x=np.linspace(0,100,10)
y=np.random.randint(0,10,10)

plt.bar(x,y,width=5,color='r')

plt.xlabel('grade') 
plt.ylabel('number of people') 
```

```python
labels='meals','medical treatments','snack','traffic','books'
spam=[0.68,0.02,0.11,0.09,0.1]
separated=(0,0,0.2,0,0)
colors=['orange','blue','yellow','green','red']
plt.pie(spam,labels=labels,colors=colors,autopct='%1.1f%%',explode=separated)
plt.axis('equal')
plt.show()
```
